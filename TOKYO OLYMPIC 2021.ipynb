We will use value_counts to get a sorted order of values so we can use a fixed number of values to demonstrate which countries most players come from. Using all the countries would make the chart too clustered.

```python
%matplotlib inline



countries_group = olympic.groupby('NOC')
countries = [NOC for NOC, olympic in countries_group]

y = olympic.NOC.value_counts().values[0:31]
x = olympic.NOC.value_counts().index[0:31]

plt.figure(figsize=(20,5))
plt.bar(x,y)
plt.xlabel("Countries")
plt.ylabel("Number of athletes")
plt.xticks(rotation='vertical',size=10)
plt.yticks(size=15)
plt.show()
```

Observation:

United States of America country there are highest number of atheletes ie. 225 who have participated in the Olympic 2021

2. Which discipline is most popular and which country has the highest participants in it?

```python
countries_group = olympic.groupby('Discipline')
countries = [Discipline for Discipline, olympic in countries_group]

y = olympic.Discipline.value_counts().values[0:31]
x = olympic.Discipline.value_counts().index[0:31]

plt.figure(figsize=(20,5))
plt.bar(x,y)

for index, value in enumerate(y):
    
    plt.text(index, value, str(value),color='red',size=12,fontweight='bold')
    
plt.xlabel("Discipline")
plt.ylabel("Number of athletes")
plt.xticks(rotation='vertical',size=10)
plt.yticks(size=15)
plt.show()
```

Observation:

In the Ice Hockey Discipline and United States of America country have highest number of atheletes who have participated in the Olympic 2021
